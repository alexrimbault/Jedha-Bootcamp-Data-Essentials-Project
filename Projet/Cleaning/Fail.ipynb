```python
#IMPORTING LIBRARIES

import pandas as pd
```

```python
#IMPORTING THE TWO DATAFRAMES

df_unicorn = pd.read_csv('Unicorn_Clean.csv')
df_crunchbase = pd.read_csv('comp.csv')
```

```python
#CHECKING THE FIRST DATAFRAME

df_unicorn
```

```python
#CHECKING THE COLUMNS IN FIRST DATASET

df_unicorn.columns
```

```python
#DROPPING THE COLUMNS WHICH ARE IRRELEVANT IN FIRST DATASET

df_unicorn.drop(columns = ['Unnamed: 0','Investor 1', 'Investor 2', 'Investor 3','Investor 4'], axis=1, inplace=True)
```

```python
#CHECKING THE NEW FIRST DATASET WHICH HAVE BEEN FORMATED

df_unicorn
```

```python
#CHECKING THE TYPE OF EACH COLUMN

df_unicorn.dtypes
```

```python
#CONVERTING THE 'DATE JOINED' COLUMN FROM AN OBJECT TO A DATETIME

df_unicorn['Date Joined'] = pd.to_datetime(df_unicorn['Date Joined'])
df_unicorn.dtypes
```

```python
# KEEPING ONLY THE YEAR IN THE COLUMN 'Date Joined',
# DROPPING THE 'Date Joined' COLUMN

df_unicorn['Year Joined'] = df_unicorn['Date Joined'].dt.to_period('Y')
df_unicorn.drop(columns = ['Date Joined'],axis=1, inplace=True)
df_unicorn
```

```python
#SHIFTING THE COLUMN YEAR BACK TO ITS INITIAL POSITION

shift = df_unicorn.pop('Year Joined')
df_unicorn.insert(2, 'Year Joined', shift)
df_unicorn
```

```python
#CHECKING THE SECOND DATASET

df_crunchbase
```

```python
#CHECKING THE COLUMNS IN SECOND DATASET

df_crunchbase.columns
```

```python
#DROPPING THE COLUMNS WHICH ARE IRRELEVANT IN SECOND DATASET

df_crunchbase.drop(columns = ['permalink','homepage_url', 'country_code', 'state_code','region','city','category_list'], axis=1,inplace=True)
```

```python
#CHECKING THE NEW SECOND DATASET WHICH HAVE BEEN FORMATED

df_crunchbase
```

```python
#CHECKING THE TYPE OF EACH COLUMN

df_crunchbase.dtypes
```

```python
#CONVERTING THE 3 DATES COLUMNS FROM AN OBJECT TO A DATETIME

df_crunchbase['founded_at'] = pd.to_datetime(df_crunchbase['founded_at'], errors = 'coerce')
df_crunchbase['first_funding_at'] = pd.to_datetime(df_crunchbase['first_funding_at'], errors = 'coerce')
df_crunchbase['last_funding_at'] = pd.to_datetime(df_crunchbase['last_funding_at'])
df_crunchbase.dtypes
```

```python
# KEEPING ONLY THE YEAR IN THE DATES COLUMNS,
# DROPPING THE 3 INITIALS COLUMNS 

df_crunchbase['founded_at'] = df_crunchbase['founded_at'].dt.to_period('Y')
df_crunchbase['first_funding_at'] = df_crunchbase['first_funding_at'].dt.to_period('Y')
df_crunchbase['last_funding_at'] = df_crunchbase['last_funding_at'].dt.to_period('Y')
df_crunchbase
```

```python
#RENAMING THE COMPANY NAME COLUMN OF THE SECOND DATASET

df_crunchbase.rename(columns={"name":"Company"}, inplace=True)
```

```python
#FUNCTION ALLOWING TO CHECK THE NUMBER OF MATCHING VALUES BETWEEN THE TWO COMPANIES COLUMNS, 
#OF THE TWO DATASET

common = []

def lower(x):
    return x.lower()
for company in df_unicorn['Company'].apply(lower):
    if df_crunchbase[df_crunchbase['Company'].apply(lower).str.find(company) != -1].shape[0] != 0:
        common.append(company)
```

```python
#CHECKING THE NUMBER OF MATCHING VALUES

len(common)
```

```python
#MERGING THE TWO DATASETS

df_final = df_unicorn.merge(df_crunchbase, on="Company", how = 'inner')
```

```python
#CHECKING THE NEW AND FINAL DATASET COMING FROM THE MERGE

df_final
```
